## Задача 2.1

### Прямоугольное окно

```python
N = 5
win = sci.signal.windows.boxcar(20 + N)
plt.stem(win)
```

```python
N_d = 4096
spec = np.abs(fft.fft(win, n=N_d))
plt.plot(spec)
```

```python
w = np.shape(np.where(spec >= spec[0]/2))[1] / 2 # Half width of the main peak
dnu = w / N_d
print(f"Half width = {dnu} nu0")
```

```python
spec /= np.max(spec)
```

```python
peaks = [0]
peaks = np.append(peaks, sci.signal.find_peaks(spec)[0])
plt.plot(spec, 'r')
plt.stem(peaks, spec[peaks])
```

```python
peaksdb = 10 * np.log10(spec[peaks])
print("Лепестки на уровне:", np.sort(peaksdb)[-2], "дБ")
```

```python
plt.scatter(np.log2(peaks[1:]), peaksdb[1:])
# plt.plot(np.log2(np.arange(0, N_d)), 10 * np.log10(spec))
```

```python
print("Спад:", (peaksdb[2] - peaksdb[1]) / (np.log2(peaks[2]) - np.log2(peaks[1])), "дБ/октаву")
```

## Задача 2.5

```python
N = 64
N_d = 4096
win_var = sci.signal.windows.bartlett(N)

m2 = -2
m3 = 5

a1 = 10 + 5
a2 = a1 / 6
a3 = a1 / 2

def func(k):
    return a1 * np.sin(2 * np.pi * k * a1 / N) + \
    a2 * np.cos(2 * np.pi * k * (a1 + m2) / N) - \
    a3 * np.sin(2 * np.pi * k * (a1 + m3) / N)
```

```python
sig = np.fromfunction(func, (N,))
spec = np.abs(fft.fft(sig, n=N_d))
plt.plot(spec)
```

```python
win = sci.signal.windows.blackman(N)
winsig = sig * win
spec = np.abs(fft.fft(winsig, n=N_d))
plt.grid()
plt.plot(spec)
```

```python
winsig = sig * win_var
spec = np.abs(fft.fft(winsig, n=N_d))
plt.grid()
plt.plot(spec)
```

```python
win = sci.signal.windows.chebwin(N, 16)
winsig = sig * win
spec = np.abs(fft.fft(winsig, n=N_d))
plt.grid()
plt.plot(spec)
```

```python
win = sci.signal.windows.kaiser(N, 1.9)
winsig = sig * win
spec = np.abs(fft.fft(winsig, n=N_d))
plt.grid()
plt.plot(spec)
```

```python
plt.plot(sig)
```
